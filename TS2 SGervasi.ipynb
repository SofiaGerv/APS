# Introducción
El objetivo principal de este trabajo es implementar un generador de señales que permita parametrizar la amplitud máxima, el valor medio, la frecuencia, la fase, la cantidad de muestras y la frecuencia de muestreo del ADC.  Una vez desarrollado el codigo, se simulará el comportamiento de la señal a las frecuencias de 1 Hz, 500 Hz, 999 Hz, 1001 Hz y 2001 Hz, para analizar su comportamiento bajo estas condiciones.

# Experiencia I
Defino la cantidad de muestras (N) como 1000, y determino la frecuencia f0 como 1Hz. 


```python
# -*- coding: utf-8 -*-
"""
Created on Mon Mar 10 20:43:20 2025

@author: Sofía
"""

#Trabajo Semanal 1

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

#Defino variables
A= 1  #Volts
Vm = 0  #Volts
f0 = 1  #Hz
fase = 0 #Radianes
N = 1000
fs = 1000 #Hz

#Defino la funcion
def mi_funcion_sen(A, Vm, f0, fase, N, fs):
    tt= np.linspace(0,(N-1)*(1/fs),N)
    xx = A* np.sin(2 * np.pi * f0 * tt + fase) + Vm
    return xx, tt

xx, tt = mi_funcion_sen(A, Vm, f0, fase, N, fs)

# Grafico la señal
plt.title('Experiencia I')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (V)')
plt.plot(tt, xx)
   
```

```python
# Experiencia II
Luego, se procede a modificar la frecuencia (f0) a 500 Hz. 
```

```python
f0 = 500  #Hz
xx, tt = mi_funcion_sen(A, Vm, f0, fase, N, fs)

plt.title('Experiencia II')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (V)')
plt.plot(tt, xx)
   
```

El teorema de Nyquist-Shannon establece que para poder reconstruir una señal correctamente a partir de sus muestras, la frecuencia de muestreo debe ser al menos el doble de la frecuencia más alta presente en la señal original.

Para el caso de f0 = 1 Hz, la señal se reconstruye con precisión porque la frecuencia de muestreo (fs = 1000 Hz) es mucho mayor que el doble de la frecuencia de la señal original. 

Por otro lado, si f0 = 500 Hz, la frecuencia de f0 es exactamente igual a la frecuencia de Nyquist. Cuando muestreamos una señal a exactamente su frecuencia de Nyquist, cada ciclo queda representado por solo dos puntos, lo que provoca quel el gráfico se vea como una onda triangular en lugar de una sinusoide. 

En el caso de f0 = 999 Hz, la frecuencia de la señal supera la frecuencia de Nyquist. Esto da lugar a un fenómeno llamado aliasing, donde la señal reconstruida tiene una frecuencia diferente a la original. La frecuencia de la señal resultante se determina de la siguiente forma: f′ = f0 + k⋅fs

Si calculamos la frecuencia f′ de la señal reconstruida, obtenemos:
f ′= 999 − 1000 = −1Hz
Esto significa que la señal se reconstruye incorrectamente como una señal de 1 Hz pero invertida en fase. Esto se puede apreciar comparando los gráficos de las experiencias I y III. 

Finalmente, cuando f0 = 1001 Hz, la frecuencia de la señal es mayor que la frecuencia de muestreo. En este caso, la señal también se reconstruye incorrectamente como una señal de 1 Hz, aunque originalmente era de 1001 Hz. Un comportamiento similar ocurre cuando f0 = 2001 Hz.

En conclusión, la frecuencia de muestreo debe ser al menos el doble de la frecuencia máxima de la señal para asegurar una reconstrucción precisa. Cuando la frecuencia de muestreo supera la frecuencia de Nyquist, el resultado es una reconstrucción incorrecta de la señal con una frecuencia diferente a la original.
